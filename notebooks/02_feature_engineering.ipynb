Feature Engineering

Ziel: 

In diesem Notebook werden aus den historischen Premier-League-Daten aussagekräftige Merkmale (Features) erstellt.

Diese Merkmale dienen später als Eingabedaten für die Machine-Learning-Modelle (Logistische Regression und Neuronales Netz).

Feature	Beschreibung
HomeGoalsLast5:	Durchschnittliche Heimtore
AwayGoalsLast5:	Durchschnittliche Auswärtstore
HomeConcededLast5:	Durchschnittliche Heim-Gegentore
AwayConcededLast5:	Durchschnittliche Auswärts-Gegentore
HomeFormLast5:	Durchschnittliche Punkte aus den letzten 5 Heimspielen
AwayFormLast5:	Durchschnittliche Punkte aus den letzten 5 Auswärtsspielen

```python
import pandas as pd 
import os

# Tordurchschnitt der letzten 5 Spiele 

DATA_PATH = "../data/raw"

files = [f for f in os.listdir(DATA_PATH) if f.endswith(".csv")]

dfs = []

for file in files:
    df = pd.read_csv(os.path.join(DATA_PATH, file))
    df["Season"] = file.replace(",csv", "")
    dfs.append(df)

matches = pd.concat(dfs, ignore_index=True)

matches["Date"] = pd.to_datetime(matches["Date"], dayfirst=True)

matches = matches.sort_values("Date")

matches.reset_index(drop=True, inplace=True)

matches.head
```

Feature 1: Heimtore im aktuellen Spiel

```python
matches[["HomeTeam", "FTHG"]].head()
```

```python
# Durschnitttlih erzielte Tore der Heimmannschaft in den letzten 5 Spielen

matches["HomeGoalsLast5"] = (
    matches.groupby("HomeTeam")["FTHG"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

matches[["Date", "HomeTeam", "FTHG", "HomeGoalsLast5"]].head(15)
```

```python
# Durschnitttlih erzielte Tore der Auswärtsmannschaft in den letzten 5 Spielen

matches["AwayGoalsLast5"] = (
    matches.groupby("AwayTeam")["FTAG"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

matches[
    ["Date", "AwayTeam", "FTAG", "AwayGoalsLast5"]
].head(15)
```

```python
# Wie viele Tore kassierte das Heimteam im Schnitt in seinen letzten 5 Heimspielen?

matches["HomeConcededLast5"] = (
    matches.groupby("HomeTeam")["FTAG"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

matches[
    ["Date", "HomeTeam", "FTAG", "HomeConcededLast5"]
].head(15)
```

```python
# Wie viele Tore kassierte das Auswärtsteam im Schnitt in seinen letzten 5

matches["AwayConcededLast5"] = (
    matches.groupby("AwayTeam")["FTHG"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

matches[
    ["Date", "AwayTeam", "FTHG", "AwayConcededLast5"]
].head(15)
```

Form der Mannschaft

```python
# Punkte für das Heimteam 

matches["HomePoints"] = 0

matches.loc[matches["FTR"] == "H", "HomePoints"] = 3
matches.loc[matches["FTR"] == "D", "HomePoints"] = 1
matches.loc[matches["FTR"] == "A", "HomePoints"] = 0

matches[["HomeTeam", "FTR", "HomePoints"]].head(10)
```

```python
# Punkte für das Auswärtsteam

matches["AwayPoints"] = 0

matches.loc[matches["FTR"] == "A", "AwayPoints"] = 3
matches.loc[matches["FTR"] == "D", "AwayPoints"] = 1
matches.loc[matches["FTR"] == "H", "AwayPoints"] = 0

matches[["AwayTeam", "FTR", "AwayPoints"]].head(10)
```

```python
# Form der letzten 5 Heimspiele

matches["HomeFormLast5"] = (
    matches.groupby("HomeTeam")["HomePoints"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

matches[
    ["HomeTeam", "HomePoints", "HomeFormLast5"]
].head(15)
```

```python
# Form der letzten 5 Auswärtsspiele

matches["AwayFormLast5"] = (
    matches.groupby("AwayTeam")["AwayPoints"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

matches[
    ["AwayTeam", "AwayPoints", "AwayFormLast5"]
].head(15)
```

```python
# Splate BTTS fehlte, aus Notebook 01 nicht übernommen, jedes notebook sollte eigenständig sein.
matches["BTTS"] = (
    (matches["FTHG"] > 0) &
    (matches["FTAG"] > 0)
).astype(int)

matches["BTTS"].value_counts()
```

Features überprüfen

```python
feature_columns = [
    "HomeGoalsLast5",
    "AwayGoalsLast5",
    "HomeConcededLast5",
    "AwayConcededLast5",
    "HomeFormLast5",
    "AwayFormLast5",
    "BTTS"
]

matches[feature_columns].head(10)
```

Datensatz für das Machine Learning vorbereiten

```python
# Fehlende Werte prüfen

matches[feature_columns].isnull().sum()
```

```python
# NaN-Zeilen entfernen

matches = matches.dropna(subset=[
    "HomeGoalsLast5",
    "AwayGoalsLast5",
    "HomeConcededLast5",
    "AwayConcededLast5",
    "HomeFormLast5",
    "AwayFormLast5"
])

matches.reset_index(drop=True, inplace=True)

print(matches.shape)
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)

matches.to_csv("../data/Processed/btts_features.csv", index=False)

print ("✅ Datensatz gespeichert")
```
