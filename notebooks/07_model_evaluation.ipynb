# Model Evaluation

In diesem Notebook werden die entwickelten Machine-Learning-Modelle miteinander verglichen.

Folgende Modelle werden bewertet:

- Logistic Regression
- Neural Network V1
- Neural Network V2

Die Bewertung erfolgt anhand der Metriken Accuracy, Precision, Recall und F1-Score.

```python
import pandas as pd
```

Ergebnisse

```python
results = pd.DataFrame({

    "Model": [
        "Logistic Regression",
        "Neural Network V1",
        "Neural Network V2"
    ],

    "Accuracy": [
        0.548,
        0.591,
        0.540
    ],

    "Precision": [
        0.572,
        0.597,
        0.541
    ],

    "Recall": [
        0.904,
        0.927,
        0.929
    ],

    "F1 Score": [
        0.702,
        0.726,
        0.684
    ]

})

results
    
```

Accuracy

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Accuracy"])

plt.title("Accuracy-Comparison")

plt.show()
```

Precision

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Precision"])

plt.title("Precision Comparison")

plt.ylabel("Precision")

plt.show()
```

Recall

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["Recall"])

plt.title("Recall Comparison")

plt.ylabel("Recall")

plt.show()
```

F1 Score

```python
plt.figure(figsize=(8,5))

plt.bar(results["Model"], results["F1 Score"])

plt.title("F1 Score Comparison")

plt.ylabel("F1 Score")

plt.show()
```

## Interpretation

Die logistische Regression diente als Baseline-Modell.

Das erste neuronale Netz konnte alle Bewertungsmetriken verbessern und erzielte die beste Gesamtleistung.

Durch zusätzliche Features (Shots on Target und BTTS-Historie) konnte die Leistung im zweiten neuronalen Netz nicht weiter gesteigert werden.

Dies deutet darauf hin, dass zusätzliche Merkmale nicht automatisch zu einer besseren Vorhersage führen. Entscheidend ist die Qualität und Aussagekraft der Features.

Für zukünftige Arbeiten könnten weitere Informationen wie Heim- und Auswärtsform, Tabellenpositionen oder Expected Goals (xG) integriert werden.
