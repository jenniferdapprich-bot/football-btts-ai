```python
from tensorflow.keras.models import load_model
import joblib

model = load_model("../models/neural_network_v2.keras")
scaler = joblib.load("../models/scaler_v2.pkl")

print("Modell erfolgreich geladen!")
```

```python
import pandas as pd

matches = pd.read_csv("../data/processed/btts_features_v2.csv")

features = [
    "HomeGoalsLast5",
    "AwayGoalsLast5",
    "HomeConcededLast5",
    "AwayConcededLast5",
    "HomeFormLast5",
    "AwayFormLast5",
    "HomeShotsOnTargetLast5",
    "AwayShorsOnTargetLast5",
    "HomeBTTSLast5",
    "AwayBTTSLast5"
]
```

```python
home_team = "Liverpool"
away_team = "Arsenal"
```

Vorhersagedaten erzeugen:

```python
home = matches[matches["HomeTeam"] == home_team].iloc[-1]
away = matches[matches["AwayTeam"] == away_team].iloc[-1]

prediction = pd.DataFrame([{
    "HomeGoalsLast5": home["HomeGoalsLast5"],
    "AwayGoalsLast5": away["AwayGoalsLast5"],
    "HomeConcededLast5": home["HomeConcededLast5"],
    "AwayConcededLast5": away["AwayConcededLast5"],
    "HomeFormLast5": home["HomeFormLast5"],
    "AwayFormLast5": away["AwayFormLast5"],
    "HomeShotsOnTargetLast5": home["HomeShotsOnTargetLast5"],
    "AwayShotsOnTargetLast5": away["AwayShotsOnTargetLast5"],
    "HomeBTTSLast5": home["HomeBTTSLast5"],
    "AwayBTTSLast5": away["AwayBTTSLast5"]
}])

prediction
```

Vorhersage berechnen:


```python
prediction_scaled = scaler.transform(prediction)

probability = model.predict(prediction_scaled, verbose=0)[0][0]

print(f"Spiel: {home_team} vs. {away_team}")
print(f"BTTS-Wahrscheinlichkeit: {probability:.2%}")

if probability >= 0.5:
    print("✅ Prognose: Beide Teams treffen.")
else:
    print("❌ Prognose: Mindestens ein Team trifft nicht.")
```
